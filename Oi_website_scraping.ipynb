```python
from bs4 import BeautifulSoup
from selenium import webdriver
import requests
import pandas as pd
```

```python
url="https://www.oi.com.br/blog/entretenimento/"
```

```python
driver=webdriver.Chrome()
driver.get(url)
```

```python
soup=BeautifulSoup(driver.page_source,"html.parser")
```

```python
print(soup.prettify)
```

```python
title=soup.find("h2",{"class":"fw-bold size-font-titulo"}).text.strip()
```

```python
title
```

```python
titles=[]
for i in range(1,4):
    driver.get("https://www.oi.com.br/blog/entretenimento/?pagina={}".format(i))
    for tit in soup.find_all("h2",{"class":"fw-bold size-font-titulo"}):
        titl=tit.get_text(strip=True)
        titles.append(titl)
```

```python
titles
```

```python
len(titles)
```

```python
descr=soup.find("div",{"class":"col-12 fs-6 mb-3 noticia-cards"}).text.strip()
```

```python
descr
```

```python
Des=[]
for i in range(1,4):
    driver.get("https://www.oi.com.br/blog/entretenimento/?pagina={}".format(i))
    for des in soup.find_all("div",{"class":"col-12 fs-6 mb-3 noticia-cards"}):
        desc=des.get_text(strip=True)
        Des.append(desc)
```

```python
Des
```

```python
len(Des)
```

```python
data={"Titles":titles,"Description":Des}
```

```python
df=pd.DataFrame(data=data)
```
